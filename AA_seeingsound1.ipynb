#### **1 基于 Python 与 librosa 包，静态地看**

&emsp;

以上从 Scratch 中导出的26张波形图按如下方法拼接而成：

```python
# 导入 pillow 包
from PIL import Image   
```

```python
# 为了尽可能无缝拼接，对 26 张图片进行裁减，特别是要去掉四周的图片边框以及左右两边的留白

img_crop = []
for i in range(1,26):
    item = str(i)
    path = "img/waveplot/w" + item + ".png"
    img = Image.open(path)
    img1 = img.crop((579,500,1515,1020))  #这一步是对每一张图片做裁减，（579,500,1515,1020)是多次实验后找到的理想裁减大小，四个值分别对应左上起始和右下终止像素值
    img_crop.append(img1)

    
```

```python
# 先按拼接后图片的大小生成一个空白图片 canva，它的宽等于 26 张裁减后波形图宽度之和，高度与单张波形图高度相同

w, h = img_crop[0].size
canva = Image.new(img.mode, (w*26, h))  
```

```python
# 将 26 张单张波形图拼贴入 canva 中
i = 0
for img in img_crop:  
    canva.paste(img, box=(i*w,0))
    i = i + 1
```

```python
# 将拼贴后的图片保存到本地

canva.save('waveplot.png')
```
